```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


def model(y, t):
    ## y = [ Omega, Theta ]
    w = y[0]
    th = y[1]
    dth_dt = w
    dw_dt = (-10)*np.sin(np.deg2rad(th))
    return [dw_dt, dth_dt]


# Initial condition
w0 = -1
th0 = 0
y0 = [w0, th0]

# Times at which the solution is to be computed.
t = np.linspace(0, 130, 10000)

y = odeint(model, y0, t)

plt.subplot(211)
plt.plot(t, y[:,0], 'k:', label="Velocity")
plt.legend()
plt.xlabel('t')
plt.subplot(212)
plt.plot(t, y[:,1], 'r--', label="Angle")
plt.xlabel('t')
plt.legend()
# plt.ylabel('y')
plt.show()
```
